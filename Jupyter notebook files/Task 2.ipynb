## 10.For each quantitative attribute, calculate its average, standard deviation, minimum, mode, range and maximum values.

```python

l1 = ['age','balance','duration','pdays','previous']
for i in l1:
    print(i)
    
    print("mean: ",df[i].mean())
    print("standard deviation: ",df[i].std())
    print("minimum: ",df[i].min())
    print("maximum: ",df[i].max())
    print("mode: ",df[i].mode()[0])
    print("range: ",df[i].max() - df[i].min())
    print("--------------------")
```

## 20. Histogram for Outliers

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols].hist(figsize=(12,8))
plt.show()
```

## 21. Outlier Detection using IQR

```python
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1

outliers = ((df[num_cols] < (Q1 - 1.5 * IQR)) | (df[num_cols] > (Q3 + 1.5 * IQR)))
outliers.sum()
```
