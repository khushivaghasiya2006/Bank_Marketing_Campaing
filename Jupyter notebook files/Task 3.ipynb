## TASK 3 :  Model Creation


```python
import pandas as pd
import numpy as np

df = pd.read_csv('Bank_train1.csv', sep=',')
df.columns = df.columns.str.strip()
print(df.columns)
df['y'] = df['y'].map({'yes': 1, 'no': 0})

X = df[['age', 'balance', 'duration', 'campaign']].values
y = df['y'].values

X = (X - X.mean(axis=0)) / X.std(axis=0)
X = np.c_[np.ones(X.shape[0]), X]

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def train_logistic_regression_sgd(X, y, lr=0.01, epochs=10):
    weights = np.zeros(X.shape[1])
    n_samples = X.shape[0]

    for epoch in range(epochs):
        indices = np.random.permutation(n_samples)
        X_shuff = X[indices]
        y_shuff = y[indices]

        for i in range(n_samples):
            xi = X_shuff[i].reshape(1, -1)
            yi = y_shuff[i]

            z = np.dot(xi, weights)
            pred = sigmoid(z)
            error = pred - yi

            weights -= lr * (xi.T * error).flatten()

    return weights

def predict(X, weights):
    probs = sigmoid(np.dot(X, weights))
    return (probs >= 0.5).astype(int)

weights = train_logistic_regression_sgd(X, y)
y_pred = predict(X, weights)

accuracy = np.mean(y_pred == y)
print("Model Accuracy (using SGD):", accuracy)
```
